# 2. Data structure (Series, Data Frame)

```python
# Series : 1D array

ser1 = pd.Series()

print(ser1)
```

```python
# create a series from list

Salary = [50, 25, 75, 15, 32, 90, 100, 200, 300, 500]

Salary_series = pd.Series(Salary)
print(Salary_series)
```

```python
type(Salary_series)
```

```python
# Series() don't support the Negative indexing
# Salary_series[-1]
```

```python
# Series from the Tuple

X_names = ("Chinki", "Minki", "Tinki", "Pinki")

Salary_series = pd.Series(X_names, index = ['A', 'B', 'C', 'D'])
print(Salary_series)
```

## Data frame in pandas

```python
# dataframe is 2D array / Tables

df1 = pd.DataFrame()

print(df1)
```

```python
# creating df from list
fruits = ["Arati Pandu", "Mamidi Pandu", "Kamala Pandu", "Panasa pandu"]

df1 = pd.DataFrame(fruits, columns = ['Pandu Peru'])

print(df1)
```

```python
type(df1)
```

```python
# list of list into dataframe
fruits = [["Arati Pandu",70], ["Mamidi Pandu",800], ["Kamala Pandu", 300], ["Panasa pandu",120]]

df1 = pd.DataFrame(fruits, columns = ["Pandu Peru", "Darah"], index = ['A', 'B', 'C', 'D'])
print(df1)
```

```python
# tuple of tuple into dataframe
fruits = {("Arati Pandu",70), ("Mamidi Pandu",800), ("Kamala Pandu", 300), ("Panasa pandu",120)}

df1 = pd.DataFrame(fruits, columns = ["Pandu Peru", "Darah"])
print(df1)
```

```python
# Creating a Dataframe from the Dictionary

South_Hero = {
    "Hero Name" : ["NTR", "Prabhas", "Anudeep", "BoB", "PK", "Kirti Suresh"],
    "Movie Fee" : [200, 1000, 0, 500, 50, 100]    
}

df1 = pd.DataFrame(South_Hero)
print(df1)
```

# 5. Data filtering and indexing (loc, iloc, where()..)

```python
# Query: filter all the records where salary > 100000

df[df.salary > 100000]
```

```python
# Query: filter all the records where salary > 100000 | salary < 50000

df[(df.salary > 100000) | (df.salary < 50000) ]
```

```python
# Query: Find the person record who is getting maximum salary

df[df.salary == df.salary.max()]
```

```python
df.salary.max()
```

```python
# Query: find the person record who is getting maximum proposal.

df[df.Marriage_proposal_count == df.Marriage_proposal_count.max()]
```

```python
# Query : Find the person PROFESSION who is getting minimum salary
df.Profession[df.salary == df.salary.min()]
```

```python
df[['Profession','salary']][df.salary == df.salary.min()]
```
