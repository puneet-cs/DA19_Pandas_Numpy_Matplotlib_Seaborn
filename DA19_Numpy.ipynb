# 3. Specific type of Numpy array

```python
import numpy as np
```

```python
# zeros() numpy function create an 1D array initialize with zeros.
arr = np.zeros(5)
print(arr)
```

```python
arr = np.zeros(10)
print(type(arr))
```

```python
# 2D Zero array
arr_2d = np.zeros((2,4))
print(arr_2d)
```

```python
# 3D Zero array
arr_3d = np.zeros((2,3,4))
print(arr_3d)
```

```python
# np.ones() : array initialize with ones
ones_1d = np.ones(6)
print(ones_1d)
```

```python
# np.ones() : array initialize with ones
ones_2d = np.ones((2,4))
print(ones_2d)
```

```python
# np.ones() : array initialize with ones
ones_3d = np.ones((2,4,3))
print(ones_3d)
```

```python
# np.full() : fill the array with any number

arr_full = np.full(10, 500)
print(arr_full)
```

```python
arr_full = np.full((3,2,4), 14)
print(arr_full)
```

```python
# identity matrix : all diagnol elements are 1.
# identity matrix is a Squared matrix
id_arr = np.eye(5)
print(id_arr)
```

```python
# numpy random package

# rand() : random values between 0 and 1

np.random.rand(10)  
```

```python
# 2D random array
np.random.rand(2,4,2)
```

```python
# uniform distribution: every value have equal chance to get selected
np.random.uniform(low = 1, high = 15, size = (2,3,4))
```

```python
# here n means Normal distribution : most of the values are between -3 to +3
np.random.randn(4)
```

```python
# create a 2D array which is having normal distributed data

np.random.randn(3,4)
```

```python
np.random.randint(50, 70, (2,3))
```
